```python
variables_numericas = ["Elapsed_Time", "Age"]

for columna in variables_numericas:
    media = df[columna].mean()
    desviacion = df[columna].std()

    limite_inferior = media - 2 * desviacion
    limite_superior = media + 2 * desviacion

    outliers = df[
        (df[columna] < limite_inferior) |
        (df[columna] > limite_superior)
    ]

    print("Variable:", columna)
    print("Media:", round(media, 2))
    print("Desviación estándar:", round(desviacion, 2))
    print("Límite inferior:", round(limite_inferior, 2))
    print("Límite superior:", round(limite_superior, 2))
    print("Cantidad de valores atípicos:", len(outliers))
    print("-" * 40)
```

```python
"Selección de atributos para reglas de asociación"
columnas_asociacion = [
    "Family",
    "Hobbies",
    "Social_Club",
    "Political",
    "Professional",
    "Religious",
    "Support_Group"
]
```

```python
"Creación del dataset de asociación"
df_asociacion = df[columnas_asociacion].copy()

df_asociacion.head()
```

```python
"Validación de valores binarios"
for columna in columnas_asociacion:
    print(columna, df_asociacion[columna].unique())
```

```python
"Conversión de 0 y 1 a valores booleanos"
df_asociacion_bool = df_asociacion.astype(bool)

df_asociacion_bool.head()
```

```python
"Verificación de tipos finales"
df_asociacion_bool.dtypes
```

```python
"Frecuencia de pertenencia a cada organización"
frecuencias = df_asociacion.sum().sort_values(ascending=False)

frecuencias
```

```python
"Porcentaje de participación por organización"
porcentajes = (df_asociacion.mean() * 100).sort_values(ascending=False)

porcentajes.round(2)
```

```python
"Revisión preliminar de combinaciones frecuentes"
for grupo_a, grupo_b in combinations(columnas_asociacion, 2):
    soporte = (df_asociacion_bool[grupo_a] & df_asociacion_bool[grupo_b]).mean()

    if soporte > 0.10:
        print(grupo_a, "+", grupo_b, "=", round(soporte, 4))
```
